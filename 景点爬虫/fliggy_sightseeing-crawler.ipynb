```python
import re
import os
import csv
import time
import json
import requests
from bs4 import BeautifulSoup
import urllib3.contrib.pyopenssl
from urllib import request, error
from urllib.error import HTTPError,URLError
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
```

```python
def get_html(url):
    urllib3.contrib.pyopenssl.inject_into_urllib3()
    resp = requests.get(url=url, headers={'Connection':'close'},stream=True, verify=False)
    time.sleep(5)
    requests.adapters.DEFAULT_RETRIES = 5 
    if resp.status_code == 200:
        resp.encoding = "utf-8"
        return (resp.text)
    else:
        None
```

```python
with open('short_provinces.json') as json_file:
    data = json.load(json_file)
    keywords = []
    for i in data:
        keywords.append(i['name']) 
```

```python
pages = list(range(1,6))
```

```python
with open('Fliggy.csv', 'w') as f:
    # CSV
    thewriter = csv.writer(f)
    thewriter.writerow(['names','province','addresses', 'counts'])
    # Index
    for keyword in keywords:
        for page in pages:
            url = f'https://travelsearch.fliggy.com/index.htm?searchType=product&keyword={keyword}&pagenum={pages}'
            html = get_html(url)
            bs = BeautifulSoup(html, 'html.parser')
            contents = bs.find_all('div',{'class':'product-middle'})
            # 
            links = []
            for content in contents:
                links.append(content.a.attrs['href'])

            for link in links:
                html = get_html(link)
                bs = BeautifulSoup(html, 'html.parser')
                try:
                    sell_count = bs.find('dl',{'class':'sell-count'})
                    detail = bs.find_all('dl',{'class':'scenic-detail-info clearfix'})
                    # 
                    count = sell_count.find('em').text
                    info = detail[3].a.attrs
                    # 
                    name = info.get('data-title')
                    # 
                    address = info.get('data-address')
                except:
                    pass
                else:
                    # CSV
                    #thewriter.writerow([name,keyword,address,count])
                    print(':'+name, ':'+keyword,address,count)
                time.sleep(5)
```
