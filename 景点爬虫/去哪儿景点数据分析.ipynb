```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("QuNaEr.csv") 
data = data.set_index('names')
data[-5:]
```

```python
data.sort_values(by=['counts'] ,inplace=True, ascending=False)
data[:10]
```

```python
data[:10].plot(kind='bar')
plt.xlabel("景点名称")
plt.ylabel("月销量")
plt.title("月销量前十景点")
```

```python
data.head()
```

```python
new_data = data.groupby(['province'], as_index=False).sum()
new_data.head()
```

```python
new_data.sort_values(by=['counts'] ,inplace=True, ascending=False)
new_data[:10]
```

```python
data_province = new_data['province'].to_list()
data_counts = new_data['counts'].to_list()
list_df = zip(data_province,data_counts)
print(list(list_df))
```

```python
new_data.plot(kind='bar')
plt.xlabel("省份")
plt.ylabel("月销量")
plt.title("旅游省份")
```
